```python
import pandas as pd
import nltk
from sklearn.feature_extraction.text import TfidfVectorizer
documents = []
files = [ 'AHOTA.md', 'ctang7', 'djshouse.md', 'JDUGGAN1.md', 'JWILL221.MD', 'tbwhite2', 'rjlusk10.md', 
   'jalexiad.md', 'markdcarringer.md', 'deytia.md', 'kagrawa1.md', 'rcm684', 'CDAFFRON.md', 'dmercer2.md',
   'AmreenS.md', 'auxiliary.md', 'ccrumpton.md' ]
for f in files:
  a = open (f)
  str = a.read()
  documents .append(str)

#Do tf.idf magic 
tfidf = TfidfVectorizer().fit_transform(documents)
# no need to normalize, since Vectorizer will return normalized tf-idf
# we may want to see these, e.g., analyze in R
#pd .DataFrame.from_records(tfidf.A).to_csv('tfidf.csv')
```

```python
# Lets see the similarity matrix
# first create labels
idx = {}
for i in range(len(files)):
    idx[i] = files [i]

# Multiply the tfidf matrix to get all pairwise distances    
df = pd. DataFrame.from_records((tfidf * tfidf.T).A,columns=files)

#do pretty printing
pd.options.display.float_format = '{:20,.3f}'.format
df = df.rename(index=idx)
#Output to a file in case we want to look at it separately
#df.to_csv("dist.csv")
df
```

```python
from scipy.cluster.hierarchy import linkage, ward, dendrogram
from scipy.spatial.distance import pdist
import matplotlib.pyplot as plt
import numpy as np
import re

%matplotlib inline
# Clean up labels (remove .md at the end)
flab = files
for i in range(len(files)):
    flab[i] = re.sub(r"\.[mM][dD]$", "", files[i])

# Calculate cosine distance (clustering need sdistance, not similarity)   
dist1 = pdist(tfidf.A, 'cosine')
# Do clustering
linkage_matrix = linkage(dist1,method='complete')

# Plot results
plt.figure(1, figsize=(22, 4))
plt.clf()
ddata = dendrogram(linkage_matrix,
               color_threshold=1,
               p=17,labels=flab,
               truncate_mode='lastp',
               show_leaf_counts=True,
               )
plt.title("Clustering of Descriptions")
```
